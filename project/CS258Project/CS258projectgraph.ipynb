```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV files
received_seq_no_data = pd.read_csv("Received_Seq_No.csv", header=None)
window_size_data = pd.read_csv("window_size.csv", header=None)
dropped_packets_data = pd.read_csv("dropped_packets.csv", header=None)

# Plot the graph for Received_Seq_No.csv
plt.figure(figsize=(10, 5))
plt.plot(received_seq_no_data[1], received_seq_no_data[0], linestyle='-', linewidth=2)
plt.title("Received Sequence Numbers")
plt.xlabel("Time")
plt.ylabel("Sequence Number")
plt.xlim(0, 100000) # Set the x-axis range
plt.show()

# Plot the graph for window_size.csv
plt.figure(figsize=(10, 5))
plt.plot(window_size_data[1], window_size_data[0], linestyle='-', linewidth=2)
plt.title("Window Size")
plt.xlabel("Time")
plt.ylabel("Window Size")
plt.xlim(0, 1000) # Set the x-axis range
plt.show()

# Plot the graph for dropped_packets.csv
plt.figure(figsize=(10, 5))
plt.plot(dropped_packets_data[1], dropped_packets_data[0], linestyle='-', linewidth=2)
plt.title("Dropped Packets")
plt.xlabel("Time")
plt.ylabel("Dropped Sequence Number")
plt.xlim(0, 100000)
plt.ylim(0, 100000)
plt.show()
```
